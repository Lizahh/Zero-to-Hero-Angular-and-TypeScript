> * Use an interface when you want to enforce a structure for an object or class without providing implementation.

> * Use a class when you want to create objects, define properties and methods, and provide full implementation.

We can define **static properties/variables** as well.

Normal variables can only be used after creating a new object. Means without objects, you cannot access normal variables e.g. in above example, name is normal variable. Without "let member = new MyTeamMembers();" name cannot be accessed.

But we can access static variables/methods without even creating an object. e.g.

```
  class Car {
    static numberOfWheels: number = 4;
  
    brand: string;
  
    constructor(brand: string) {
      this.brand = brand;
    }
  
    getCarDetails() {
      return `This car is a ${this.brand}.`;
    }
  }
  
  # Accessing and printing static variable directly without even creating objects

  console.log(Car.numberOfWheels);

  # To access non static, must instantiate class first with object
  
  let myCar = new Car("Toyota");
  
  console.log(myCar.getCarDetails());
```

* We can assign default values to some variables as well:


```
class Car {

    brand: string = "mycar";
    model: string;
  
    constructor(model: string) {
      this.model = model;
    }
  
    cardetails() {
          // Use backticks (`) for string interpolation
    console.log(`This car is of brand ${this.brand} and of model ${this.model}`);
    }
  
  }
  
  let newcar = new Car("Corolla");
  newcar.cardetails();
```

This way we have just given one value in object and same that one value in cosntructor.

* Sometimes, we want that if the object doesn't pass value of a variable then use the default value but if object gives value then use that value. In that case, you need to define default value in constructor. To let the cals



```
class Car {
  brand: string;
  model: string;

  // Set a default value for brand in the constructor
  constructor(brand: string = "mycar", model: string) {
    this.brand = brand;
    this.model = model;
  }

  cardetails() {
    // Use backticks (`) for string interpolation
    console.log(`This car is of brand ${this.brand} and of model ${this.model}`);
  }
}

let newcar1 = new Car("Toyota", "Corolla");  // Provide both brand and model
newcar1.cardetails();  // Outputs: This car is of brand Toyota and of model Corolla

let newcar2 = new Car(undefined, "Civic");  // Only provide model, use default brand
newcar2.cardetails();  // Outputs: This car is of brand mycar and of model Civic

```
To let the class know that you have to use default value, you need to set that value as "undefined" in object.

* Must declare constructor to call automatically when an object will be created. It will be defined with values given in object.



```
class Car {

    brand: string;
    model: string;
  
    constructor(brand: string, model: string) {
      this.brand = brand;
      this.model = model;
    }
  
    cardetails() {
        // Use backticks (`) for string interpolation
        console.log(`This car is of brand ${this.brand} and of model ${this.model}`);
    }
  
  }
  
  let newcar = new Car("Toyota", "Corolla");
  newcar.cardetails();


```


* We can declare static functions as well means we can call them without creating an object:

```
  class Car {
  static numberOfWheels: number = 4;

  brand: string;
  model: string;

  constructor(brand: string, model: string) {
    this.brand = brand;
    this.model = model;
  }

  // Static method (can only access static properties)
  static getNumberOfWheels() {
    return `A car typically has ${Car.numberOfWheels} wheels.`;
  }

  // Instance method (can access instance properties)
  getCarDetails() {
    return `This car is a ${this.brand} ${this.model}.`;
  }
}

// Accessing the static property directly from the class
console.log(Car.numberOfWheels);  // Outputs: 4

// Accessing the static method directly from the class
console.log(Car.getNumberOfWheels());  // Outputs: A car typically has 4 wheels.

// Creating an instance (object) of the class
let myCar = new Car("Toyota", "Corolla");

// Accessing the instance properties and methods
console.log(myCar.getCarDetails());  // Outputs: This car is a Toyota Corolla.

```



# **C) Assigning Values to Static Variables/Methods:**



## **1. Direct Assignment Outside the Constructor:**

You can assign value to static variable directly as:

```
class Car {
  static numberOfWheels: 4;  
```

OR


You can assign a value to the static variable directly, outside of the class, after class has been declared.

```
class Car {
  static numberOfWheels: number;  // No initial value

  brand: string;
  model: string;

  constructor(brand: string, model: string) {
    this.brand = brand;
    this.model = model;
  }

  static getNumberOfWheels() {
    return `A car typically has ${Car.numberOfWheels} wheels.`;
  }
}

// Assign a value to the static variable outside the class
Car.numberOfWheels = 4;  // Directly assigning value

console.log(Car.getNumberOfWheels());  // Outputs: A car typically has 4 wheels.

```
## **2. Static Block (since ES2022)**

Another way is to use a static block to initialize or modify static properties. This is useful if the initialization involves more complex logic.

```
class Car {
  static numberOfWheels: number;

  brand: string;
  model: string;

  constructor(brand: string, model: string) {
    this.brand = brand;
    this.model = model;
  }

  static getNumberOfWheels() {
    return `A car typically has ${Car.numberOfWheels} wheels.`;
  }

  // Static block to initialize or modify static variables
  static {
    Car.numberOfWheels = 4;  // Assigning value inside a static block
  }
}

console.log(Car.getNumberOfWheels());  // Outputs: A car typically has 4 wheels.

```




* Normal variables are called instance variables.
* Static variables are called static variables.

* Normal methods are called instance methods. because they are dependent on instances/objects to be created to be called.

* Static methods are called static methods.
