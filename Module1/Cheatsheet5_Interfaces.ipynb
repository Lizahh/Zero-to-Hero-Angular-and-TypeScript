# **Interfaces:**

An interface is like a contract/blueprint that defines the structure of an object. It tells TypeScript what properties and methods an object should have, but does not provide the actual implementation.

So we can define interfaces as:



```
interface namesConvention {
    name: string;
}
function printMyName(namescollection: namesConvention) {
    console.log(namescollection.name);
}
printMyName({name: "aliza"})
printMyName({label: "aliza"})

```

Now when you will try to compile ts code to js it will raise and error i.e.

image.png



So, we get to know that what we are doing wrong.

Now if we add another property "age" in interface and do not use it in function it will raise an error:



```
interface namesConvention {
    name: string;
    age: number;
}
function printMyName(namescollection: namesConvention) {
    console.log(namescollection.name);
}
printMyName({name: "aliza"}
```
image.png




So, if we want to keep things optional in interface we should use ? with that property name as:

age?


```
interface namesConvention {
    name: string;
    age?: number;
}
function printMyName(namescollection: namesConvention) {
    console.log(namescollection.name);
}
printMyName({
  name: "aliza"
  })
```

Now it will compile perfectly.




# **Why use an interface?**

* **To define shapes:** Use interfaces to define what an object should look like, so you know exactly what properties and methods it must have.


* **To ensure consistency:** If different objects should have the same structure, an interface makes sure they all have the same properties and methods.

* **To help with type checking:** TypeScript will check that objects follow the structure defined by the interface.

> Example:



```
interface Person {
    name: string;
    age: number;
    greet(): void;  // This means the object must have a `greet` method
  }

let person1: Person = {
    name: "Aliza",
    age: 25,
    greet() {
      console.log("Hello, my name is " + this.name);
    }
  };
  
let person2: Person = {
    name: "John",
    age: 30,
    greet() {
      console.log("Hello, I'm " + this.name);
    }
  };

```




# **Advanced Concepts - Interfaces:**

## **1. Extending Interfaces:**

You can extend an interface from another interface to build on existing definitions. This is useful when you have a base structure that other interfaces share.



```
interface Person {
  name: string;
  age: number;
}

interface Employee extends Person {
  employeeId: number;
}

let employee1: Employee = {
  name: "John",
  age: 30,
  employeeId: 101
};
```



In this example, the Employee interface extends the Person interface, meaning it inherits all the properties of Person (name and age), plus it adds employeeId.

## **2. Readonly Properties:**

You can mark properties in an interface as readonly so that once a value is set, it cannot be changed like you do set values of tuple once in python that cannot be changed.



```
# declaration

interface Person {
    readonly id: number;
    name: string;
  }
# Initialization
  
  let person1: Person = {
    id: 1,
    name: "Aliza"
    };.

# Changing values

person1.name = "Sarah";  // Allowed
person1.id = 2;  // Error: id is readonly
  
```

This is useful when you want to ensure certain properties remain constant after they are initialized.





## **3. Function Types in Interfaces:**

You can define the type of a function inside an interface, including the parameters and return type.



```
interface MathOperation {
  (x: number, y: number): number;
}

let add: MathOperation = (a, b) => a + b;
let subtract: MathOperation = (a, b) => a - b;

```
Here, MathOperation defines the type of any function that takes two numbers and returns a number.




## **5. Indexable Types:**

You can use interfaces to describe objects that have dynamic or unknown keys, like dictionaries, using index signatures.



```
interface StringDictionary {
  [key: string]: string;
}

let myDictionary: StringDictionary = {
  firstName: "Aliza",
  lastName: "Smith"
};
```
Here, the interface allows for any property with a string key and string value.


## **6. Intersection Types:**

TypeScript also supports combining multiple interfaces into a single type using intersection types.


```
interface Person {
  name: string;
}

interface Address {
  street: string;
}

type PersonWithAddress = Person & Address;

let person: PersonWithAddress = {
  name: "Aliza",
  street: "Main St"
};

```

## **7. Implementing Interfaces in Classes:**

You can also use interfaces to define the structure that classes should follow.



```
interface Person {
  name: string;
  greet(): void;
}

class Employee implements Person {
  name: string;
  constructor(name: string) {
    this.name = name;
  }

  greet() {
    console.log("Hello, my name is " + this.name);
  }
}

let employee1 = new Employee("Aliza");
employee1.greet();

```
